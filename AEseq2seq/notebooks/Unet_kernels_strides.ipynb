```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def create_hyperparameter_comparison(analysis_df):
    """
    Creates visualizations to compare the impact of different hyperparameters
    """
    
    # Create figure for hyperparameter analysis
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    
    sns.scatterplot(data=analysis_df, x='arc_kernel', y='valid_Accuracy', hue='arc_skip_conn', ax=axes[0,1])
    axes[0,1].set_title('Accuracy vs Kernel Size')
    
    sns.scatterplot(data=analysis_df, x='arc_skip_conn', y='valid_Accuracy',hue='arc_kernel', ax=axes[0,2])
    axes[0,2].set_title('Accuracy vs Skip Connections')
    
    sns.scatterplot(data=analysis_df, x='arc_stride_1', y='valid_Accuracy', 
                    size='arc_stride_2', hue='arc_skip_conn', ax=axes[1,0])
    axes[1,0].set_title('Accuracy vs Strides')
    
    sns.scatterplot(data=analysis_df, x='arc_filters', y='valid_Accuracy',
                    size='arc_rank', hue='arc_kernel', ax=axes[1,1])
    axes[1,1].set_title('Accuracy vs Filters and Rank')
    
    sns.boxplot(data=analysis_df, x='arc_stride_1', y='valid_Accuracy', hue='arc_stride_2', ax=axes[1,2])
    axes[1,2].set_title('Accuracy vs Stride 1')
    
    plt.tight_layout()
    plt.show()

# Función para mostrar las mejores configuraciones
def show_best_configs(params_df, test_df, top_n=5):
    """
    Shows the best performing configurations based on test accuracy
    """
    analysis_df = pd.merge(params_df, test_df[['run_uuid', 'test_Accuracy', 'test_F1']], on='run_uuid')
    
    # Sort by test accuracy
    best_configs = analysis_df.sort_values('test_Accuracy', ascending=False).head(top_n)
    
    # Select relevant columns
    columns = ['name', 'arc_filters', 'arc_kernel', 'arc_stride_1', 'arc_stride_2', 
              'arc_skip_conn', 'test_Accuracy', 'test_F1']
    
    return best_configs[columns]

# Ejemplo de uso:
# create_loss_curves(train_df, params_df)
# create_hyperparameter_comparison(params_df, test_df)
# print(show_best_configs(params_df, test_df))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_training_curves(train_df, params_df):
    # Configuración general de estilo
    sns.set_style("whitegrid")
    plt.rcParams['figure.figsize'] = (15, 10)
    
    
    # Verificar columnas disponibles para el gráfico de líneas
    metrics_to_plot = [col for col in ['train_loss', 'train_Accuracy', 'valid_loss', 'valid_Accuracy'] 
                      if col in train_df.columns]
    
    # Agrupar por nombre del modelo y época
    grouped = train_df.groupby(['name', 'step'])
    
    # Plot de métricas disponibles
    plot_positions = {
        'train_loss': (0,0),
        'valid_loss': (0,1),
        'train_Accuracy': (1,0),
        'valid_Accuracy': (1,1)
    }
    
    plot_titles = {
        'train_loss': 'Train Loss vs Epoch',
        'valid_loss': 'Validation Loss vs Epoch',
        'train_Accuracy': 'Train Accuracy vs Epoch',
        'valid_Accuracy': 'Validation Accuracy vs Epoch'
    }
    
    
    # 2. Gráfico de barras comparando métricas finales
    metrics_final = [col for col in params_df.columns 
                    if any(col.startswith(prefix) for prefix in ['train_', 'valid_', 'test_'])
                    and col not in ['train_F1', 'valid_F1', 'test_F1']]
    
    n_metrics = len(metrics_final)
    n_cols = 3
    n_rows = (n_metrics + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5*n_rows))
    fig.suptitle('Comparación de métricas finales por modelo', fontsize=16)
    
    for idx, metric in enumerate(metrics_final):
        ax = axes[idx//n_cols, idx%n_cols] if n_rows > 1 else axes[idx]
        sns.barplot(data=params_df, x='name', y=metric, ax=ax)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
        ax.set_title(f'Final {metric}')
    
    # Ocultar subplots vacíos
    for idx in range(len(metrics_final), n_rows * n_cols):
        if n_rows > 1:
            axes[idx//n_cols, idx%n_cols].set_visible(False)
        else:
            axes[idx].set_visible(False)
    
    plt.tight_layout()
    
    # 3. Heatmap de correlación entre hiperparámetros y métricas finales
    hyperparams = [col for col in ['arc_kernel', 'arc_stride_1', 'arc_stride_2', 'arc_skip_conn']
                   if col in params_df.columns]
    
    metrics = [col for col in metrics_final if col in params_df.columns]
    
    if hyperparams and metrics:
        correlation_matrix = params_df[hyperparams + metrics].corr()
        
        plt.figure(figsize=(15, 6))
        sns.heatmap(correlation_matrix.loc[hyperparams, metrics], 
                    annot=True, 
                    cmap='coolwarm', 
                    center=0,
                    fmt='.2f')
        plt.title('Correlación entre hiperparámetros y métricas')
        plt.tight_layout()

# Uso de la función
plot_training_curves(train, params)
```
