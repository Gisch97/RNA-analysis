```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('batch_test.csv')
```

```python
fig, ax = plt.subplots(2,1, figsize= (6,6), sharex=True) 
f =sns.scatterplot(data=df,
                x='batch_size',
                y='test_loss',
                ax=ax[0])
f.grid()
g =sns.scatterplot(data=df,
                x='batch_size',
                y='test_Accuracy',
                ax=ax[1])
g.grid()
fig.suptitle('Metrica de test por batch size')
fig.tight_layout()
```

```python
f =sns.scatterplot(data=df,
                x='batch_size',
                y='test_F1')
f.grid()
```

```python
df['delta_acc'] = df['test_Accuracy'] - df.test_Accuracy.mean()
```

```python
fig, ax = plt.subplots()
df.delta_acc.hist(bins=30, ax=ax)
ax.set_title('Histograma de variación de Precisión')
ax.set_xlabel(r'$\Delta$ Acc')
```
