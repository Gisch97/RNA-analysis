```python
FAMILIES_PATH = '../pruebas/crossFamily/weights/'
ablations = ['C1D_C2D', 'no_ResNet2d'] 
families = os.listdir(FAMILIES_PATH)
print(f"families: {families}")
```

```python
train_logs = []
 
train_file = 'train_log.csv'
for ablation in ablations:
    for family in families:
        # Ruta del subdirectorio donde se encuentran los archivos train_log.csv
        train_file_path = os.path.join(FAMILIES_PATH, family, ablation)
        train_file_valid_path = os.path.join(FAMILIES_PATH, family, ablation, 'valid_with_test')
        print(train_file_path)
        print(train_file_valid_path)
        # Verificamos si el directorio existe
        if os.path.isdir(train_file_path):
            # Verificar si el archivo 'train_log.csv' existe en el directorio
            full_path = os.path.join(train_file_path, train_file)
            
            # Si el archivo existe, lo leemos
            if os.path.exists(full_path):
                try:
                    df = pd.read_csv(full_path)
                    
                    # Agregar columnas de 'ablation' y 'crossFamily'
                    df['ablation'] = ablation
                    df['crossFamily'] = family
                    df['valid_set'] = 'None'
                    
                    # Agregar el DataFrame a la lista
                    train_logs.append(df)
                    print(f"Successfully loaded {full_path}")
                except Exception as e:
                    print(f"Error loading {full_path}: {e}")
        
            # Verificamos si el directorio existe
            if os.path.isdir(train_file_valid_path): 
                full_path = os.path.join(train_file_valid_path, train_file)
                
                # Si el archivo existe, lo leemos
                if os.path.exists(full_path):
                    try:
                        df = pd.read_csv(full_path)
                        
                        # Agregar columnas de 'ablation' y 'crossFamily'
                        df['ablation'] = ablation 
                        df['crossFamily'] = family
                        df['valid_set'] = 'test'
                        
                        # Agregar el DataFrame a la lista
                        train_logs.append(df)
                        print(f"Successfully loaded {full_path}")
                    except Exception as e:
                        print(f"Error loading {full_path}: {e}")

# Combinar todos los DataFrames en uno solo
if train_logs:
    df = pd.concat(train_logs, ignore_index=True)
    print(f"Total rows in combined DataFrame: {df.shape[0]}")
```

```python
import pandas as pd
train_logs = []
diff_logs = []
train_file = 'train_log.csv'
for ablation in ablations:
    for family in families:
        # Ruta del subdirectorio donde se encuentran los archivos train_log.csv
        train_file_path = os.path.join(FAMILIES_PATH, family, ablation)
        train_file_valid_path = os.path.join(FAMILIES_PATH, family, ablation, 'valid_with_test')
        print(train_file_path)
        print(train_file_valid_path)
        # Verificamos si el directorio existe
        if os.path.isdir(train_file_path):
            # Verificar si el archivo 'train_log.csv' existe en el directorio
            full_path = os.path.join(train_file_path, train_file)
            
            # Si el archivo existe, lo leemos
            if os.path.exists(full_path):
                try:
                    print(f'read : {train_file_path}')
                    df1 = pd.read_csv(full_path)
                    
                    # Agregar columnas de 'ablation' y 'crossFamily'
                    df1['ablation'] = ablation
                    df1['crossFamily'] = family 
                    
                    # Agregar el DataFrame a la lista
                    train_logs.append(df1)
                    print(f"Successfully loaded {full_path}")
                except Exception as e:
                    print(f"Error loading {full_path}: {e}")
        
        # Verificamos si el directorio existe
        if os.path.isdir(train_file_valid_path): 
            full_path = os.path.join(train_file_valid_path, train_file)
            
            # Si el archivo existe, lo leemos
            if os.path.exists(full_path):
                try:
                    print(f'read : {train_file_valid_path}')
                    df2 = pd.read_csv(full_path)                
                    
                    # Agregar el DataFrame a la lista
                    train_logs.append(df2)
                    print(f"Successfully loaded {full_path}")
                except Exception as e:
                    print(f"Error loading {full_path}: {e}")
        diff = df1 - df2
        diff_logs.append(diff)
        
del(df1)
del(df2)
del(diff)
# Combinar todos los DataFrames en uno solo
if train_logs:
    df = pd.concat(train_logs, ignore_index=True)
    print(f"Total rows in combined DataFrame: {df.shape[0]}")

if diff_logs:
    df_diff = pd.concat(diff_logs, ignore_index=True)
    print(f"Total rows in combined DataFrame: {df_diff.shape[0]}")
```
