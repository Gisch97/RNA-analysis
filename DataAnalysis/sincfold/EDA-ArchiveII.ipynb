```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs_path = '../data/'
dfs_ = os.listdir(dfs_path)

dfs = {}
for df in dfs_:
    df_temp = pd.read_csv(dfs_path+df)
    dfs[df] = df_temp
    
df = dfs['ArchiveII.csv']
```

```python
df['sequence_len'] = df['sequence'].apply(len)

# Añadir columna de cantidad de pares de bases
df['num_base_pairs'] = df['base_pairs'].apply(len)

# Resumen estadístico
summary = df[['sequence_len', 'num_base_pairs']].describe()
print("Resumen estadístico de longitud y número de pares de bases:\n", summary)

# Visualización de distribución de longitud de secuencias y cantidad de pares de bases
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Distribución de longitud de secuencias
sns.histplot(df['sequence_len'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Distribución de Longitud de Secuencias')
axes[0].set_xlabel('Longitud de Secuencia')
axes[0].set_ylabel('Frecuencia')

# Distribución de número de pares de bases
sns.histplot(df['num_base_pairs'], bins=10, kde=True, ax=axes[1])
axes[1].set_title('Distribución de Número de Pares de Bases')
axes[1].set_xlabel('Número de Pares de Bases')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
 
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA, LatentDirichletAllocation
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar df
# df = pd.read_csv('path_to_your_file.csv')  # Asegúrate de cargar tu df correctamente

# Asegúrate de que el df tenga columnas 'id' y 'sequence'
# Si no las tienes, aquí un ejemplo ficticio:
# df = pd.DataFrame({'id': ['seq_1', 'seq_2'], 'sequence': ['ATCG', 'GCTA']})

# Paso 1: Extraer la familia de cada secuencia
df['family'] = df['id'].str.split('_').str[0]

# Convertir cada secuencia a un formato adecuado para NLP (tokenizar las bases)
df['sequence_text'] = df['sequence'].apply(lambda x: ' '.join(list(x)))

# Paso 2: Vectorización de cada secuencia individual con TF-IDF
vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(1, 3), max_features=1000)  # Limitar características
X = vectorizer.fit_transform(df['sequence_text'])
tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out(), index=df['id'])

# Paso 3: Análisis de Similitud de Coseno (a nivel de secuencia)
cosine_sim = cosine_similarity(X)
similarity_df = pd.DataFrame(cosine_sim, index=df['id'], columns=df['id'])

# Visualizar una submatriz de similitud de coseno (por ejemplo, primeras 20 secuencias)
plt.figure(figsize=(12, 10))
sns.heatmap(similarity_df, cmap="YlGnBu", annot=False)
plt.title('Cosine Similarity Between Sequences (Sample)')
plt.show()

# Paso 4: Clustering KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
df['cluster'] = kmeans.labels_  # Asegúrate de que 'cluster' esté correctamente asignado

# Paso 5: Reducción de Dimensionalidad para Visualización (PCA)
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(X.toarray())
reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])

# Asegúrate de que las columnas 'family' y 'cluster' estén presentes en el df
reduced_df['family'] = df['family'].values  # Convertir a valores de numpy para evitar errores
reduced_df['cluster'] = df['cluster'].values  # Convertir a valores de numpy para evitar errores

# Visualización de los clusters en el espacio reducido de PCA
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=reduced_df, palette='viridis', s=60)
plt.title('KMeans Clusters of Sequences')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()

# Visualizar PCA con familias y clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='family', data=reduced_df, palette='Set1', s=60)
plt.title('PCA of Sequence Families')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()

# Paso 6: Modelado de Tópicos (LDA)
lda = LatentDirichletAllocation(n_components=4, random_state=0)
lda.fit(X)

# Mostrar las palabras más representativas de cada tema
for i, topic in enumerate(lda.components_):
    print(f'Tema #{i + 1}:')
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])

# Fin del script
```
