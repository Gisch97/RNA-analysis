# Carga y filtro de datos

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import rna_utils as ru

CORR_PATH = "../data/seqsim_f_all.h5"
META_PATH = "../data/sources/ArchiveII_with_prob_and_motiv.csv"
```

```python
corr, meta = ru.load_and_align(corr_path=CORR_PATH, meta_path=META_PATH)
display(meta.head())
display(corr.head())
```

```python
fams = ru.build_intra_fam_corrs(corr, meta)
print(fams.keys())
print(meta.groupby("fam").count())
print(fams["tRNA"].shape)
```

```python
df = fams["telomerase"].copy()
np.fill_diagonal(df.values, 0)

display(df.max().sort_values(ascending=False))
m = df.max().max()
count = (df == m).sum(axis=1).values

count
```

```python
# 2) sub-matrices por familia
corrs = ru.build_intra_fam_corrs(corr, meta)
corrs_summary = ru.summary_table(corrs)
display(corrs_summary)
```

```python
corrs_filt400 = ru.filter_by_count_dict(corrs, data_size=400)
corrs_filt200 = ru.filter_by_count_dict(corrs, data_size=200)
```

# Guardo dfs

```python
# 1. Obtener el conjunto de todos los IDs usados en todos los DataFrames del diccionario
all_ids = set()
for df in corrs_filt400.values():
    all_ids.update(df.index)

# 2. Convertir a lista ordenada si querés controlar el orden
all_ids = sorted(all_ids)

# 3. Subseleccionar el DataFrame original
filtered_corr400 = corr.loc[all_ids, all_ids]

# 4. (Opcional) Verificar que la matriz resultante es cuadrada y simétrica
assert filtered_corr400.shape[0] == filtered_corr400.shape[1]
assert (filtered_corr400.values == filtered_corr400.values.T).all()

# 5. Ahora `filtered_corr` es el nuevo DataFrame con sólo los IDs usados
print(filtered_corr400.shape)
```

```python
filtered_corr400.to_hdf(
    "../data/seqsim_filtered400.h5", key="filtered_corr400", mode="w"
)
```
