```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import rna_utils as ru

CORR_PATH = "../data/seqsim_f_all.h5"
META_PATH = "../data/sources/ArchiveII_with_prob_and_motiv.csv"
```

```python
corr, meta = ru.load_and_align(corr_path=CORR_PATH, meta_path=META_PATH)
```

```python
fams = ru.build_inter_fam_corrs(corr, meta)
print(meta.groupby("fam").count())
print(fams.keys())
display(fams["23s"]["telomerase"])
```

```python
for f in fams.keys():
    print(6 * f"---{f}---")
    display(ru.summary_table(fams[f]))
```

```python
strength = ru.table_inter_fam(fams)
# strength = ru.table_max_inter_fam(fams)
df_strength = pd.DataFrame.from_dict(strength, orient="index")
display(df_strength)
```

```python
print(df_strength.shape)
display(df_strength.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_strength.values.flatten(), bins=30, kde=True)
plt.title("DistribuciÃ³n de similitudes inter/intra-familia")
plt.xlabel("Similitud promedio (triÃ¡ngulo superior)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df_long = df_strength.reset_index().melt(
    id_vars="index", var_name="Familia", value_name="Similitud"
)
df_long.rename(columns={"index": "CondiciÃ³n"}, inplace=True)

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_long, x="Familia", y="Similitud")
plt.xticks(rotation=45)
plt.title("DistribuciÃ³n de similitudes por familia")
plt.show()
```

```python
# Asegurarse de que el Ã­ndice y las columnas coincidan
common = df_strength.index.intersection(df_strength.columns)

# Obtener los valores diagonales
diagonal = df_strength.loc[common, common].values.diagonal()

# Crear un orden basado en la diagonal (de mayor a menor)
ordered_labels = common[np.argsort(-diagonal)]  # el "-" es para ordenar descendente

# Reordenar el DataFrame
df_sorted = df_strength.loc[ordered_labels, ordered_labels]

# Heatmap ordenado
plt.figure(figsize=(10, 8))
sns.heatmap(df_sorted, annot=True, fmt=".2f", cmap="viridis")
plt.title("Similitud promedio inter/intra familia ", fontsize=14)
plt.xlabel("Familia", fontsize=13)
plt.ylabel("Familia", fontsize=13)
plt.xticks(rotation=0, fontsize=13)
plt.yticks(rotation=0, fontsize=13)
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(df_strength.fillna(0))  # o df_strength.dropna(axis=1)

df_pca = pd.DataFrame(components, columns=["PC1", "PC2"], index=df_strength.index)

sns.scatterplot(data=df_pca, x="PC1", y="PC2")
for label in df_pca.index:
    plt.text(df_pca.loc[label, "PC1"] + 0.01, df_pca.loc[label, "PC2"] + 0.01, label)
plt.title("PCA sobre similitudes")
plt.show()
```

```python
# Creamos una copia para evitar modificar el original
df_no_diag = df_strength.copy()

# Recorremos la diagonal y seteamos esos valores a NaN
for i in df_no_diag.index:
    if i in df_no_diag.columns:
        df_no_diag.loc[i, i] = np.nan

# Ahora calculamos el promedio por columna sin contar la diagonal (que ahora es NaN)
column_means = df_no_diag.mean()

# Valores diagonales: similitud intra-familia
intra_family = pd.Series(
    {i: df_strength.loc[i, i] for i in df_strength.index if i in df_strength.columns}
)

# Juntamos todo en un DataFrame para anÃ¡lisis o grÃ¡fica
comparison = pd.DataFrame(
    {
        "Intra-familia (diagonal)": intra_family,
        "Inter-familia (sin diagonal)": column_means,
    }
).sort_values(by="Intra-familia (diagonal)")
comparison
```

```python
comparison.plot(kind="bar", figsize=(12, 6))
plt.title("Similitud intra vs inter-familia")
plt.ylabel("Similitud promedio")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
