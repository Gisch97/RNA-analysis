```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import rna_utils as ru

CORR_PATH = "../data/seqsim_f_all.h5"
CORR200_PATH = "../data/seqsim_filtered200.h5"
CORR400_PATH = "../data/seqsim_filtered400.h5"
META_PATH = "../data/sources/ArchiveII_with_prob_and_motiv.csv"
```

```python
corr, meta = ru.load_and_align(corr_path=CORR_PATH, meta_path=META_PATH)
corr200, meta200 = ru.load_and_align(corr_path=CORR200_PATH, meta_path=META_PATH)
corr400, meta400 = ru.load_and_align(corr_path=CORR400_PATH, meta_path=META_PATH)
```

```python
fams = ru.build_inter_fam_corrs(corr200, meta200)
print(meta200.groupby("fam")["sequence"].count())
# print(fams.keys())
# display(fams["23s"]["telomerase"])
```

```python
for f in fams.keys():
    print(6 * f"---{f}---")
    display(ru.summary_table(fams[f]))
```

```python
print("max")
strength_max = ru.table_max_inter_fam(fams)
df_strength_max = pd.DataFrame.from_dict(strength_max, orient="index")
display(df_strength_max)

print("mean")
strength = ru.table_inter_fam(fams)
df_strength = pd.DataFrame.from_dict(strength, orient="index")
display(df_strength)
```

```python
print("max")
print(df_strength_max.shape)
display(df_strength_max.describe())
print("mean")
print(df_strength.shape)
display(df_strength.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(16, 6), sharey=True, sharex=True)

sns.histplot(df_strength.values.flatten(), bins=30, kde=True, ax=ax[0])
sns.histplot(df_strength_max.values.flatten(), bins=30, kde=True, ax=ax[1])
ax[0].set_title("(mean)")
ax[1].set_title("(max)")
fig.suptitle("Distribución de similitudes inter/intra-familia")
ax[0].set_xlabel("Similitud promedio")
ax[1].set_xlabel("Similitud max")
ax[0].set_ylabel("Frecuencia")
ax[0].grid(True)
ax[1].grid(True)
fig.tight_layout()
plt.show()
```

```python
# Creamos una copia para evitar modificar el original
df_no_diag = df_strength_max.copy()

# Recorremos la diagonal y seteamos esos valores a NaN
for i in df_no_diag.index:
    if i in df_no_diag.columns:
        df_no_diag.loc[i, i] = np.nan

# Ahora calculamos el promedio por columna sin contar la diagonal (que ahora es NaN)
column_means = df_no_diag.mean()

# Valores diagonales: similitud intra-familia
intra_family = pd.Series(
    {
        i: df_strength_max.loc[i, i]
        for i in df_strength_max.index
        if i in df_strength_max.columns
    }
)

# Juntamos todo en un DataFrame para análisis o gráfica
comparison = pd.DataFrame(
    {
        "Intra": intra_family,
        "Inter": column_means,
    }
).sort_values(by="Inter")
comparison
```

```python
comparison.plot(kind="bar", figsize=(8, 6))
plt.title("Similitud intra vs inter-familia - max")
plt.ylabel("Similitud maxima")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
